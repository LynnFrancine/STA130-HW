```python
Question 1 :
The Standard Error of the mean, defined as the standard deviation of the distribution of bootstrapped means and the Standard
deviation are both standard deviations, that is they are measurements of the variability and the spread of data. However, 
they differ in that:
-The Standard deviation: works with a single sample dataset. It measures how spread each data value is from the the sample's
mean and how much each data value differs from the sample's mean. Whereas,
-The Standard Error of the mean: works with the distribution of the collection of multiple sample means of a dataset
(bootstrapped dataset). It measures how much the original sample mean differs from the population mean (which is just the
mean from the distribution of the collection of multiple bootstrapped sample means).
```

```python
Sure, here's a summary of our conversation:

1. **Difference Between Standard Error of the Mean and Standard Deviation**:
   - **Standard Deviation (SD)**: Measures the spread of individual data points around the mean. It indicates the overall variability within the dataset.
   - **Standard Error of the Mean (SEM)**: Measures the precision of the sample mean as an estimate of the population mean. It is calculated as the standard deviation divided by the square root of the sample size.

2. **Process of Calculating SEM**:
   - **Step 1**: Collect your sample data.
   - **Step 2**: Calculate the sample mean.
   - **Step 3**: Calculate the standard deviation of the sample.
   - **Step 4**: Calculate the SEM using the formula: $$SEM = \frac{SD}{\sqrt{n}}$$.

3. **Interpretation of SEM**:
   - The SEM indicates the accuracy of the sample mean as an estimate of the population mean. A smaller SEM suggests higher precision.
   - In your example, with a sample size of 5 and an SEM of approximately 0.71, it means that the sample mean is expected to vary by about 0.71 units from the true population mean.

4. **Creating a 95% Confidence Interval Using SEM**:
   - **Step 1**: Calculate the sample mean.
   - **Step 2**: Calculate the SEM.
   - **Step 3**: Determine the critical value (1.96 for 95% confidence).
   - **Step 4**: Calculate the margin of error: $$\text{Margin of Error} = 1.96 \times \text{SEM}$$.
   - **Step 5**: Construct the confidence interval: $$\text{Mean} \pm \text{Margin of Error}$$.
   - Example: For a sample mean of 7 and SEM of 0.71, the 95% confidence interval is approximately [5.61, 8.39].

This summary should cover the key points of our discussion. If you need any more details or have further questions, feel free to ask!
```

```python
Question 6 :
What is the process of bootstrapping?
-Bootstrapping is a time and cost-saving method of using data from a sample to estimate the possible data and statistics of 
its population. It consists of using a sample and creating a bunch of other samples form the original sample by using 
sampling with replacement (which implies including one data value from the original more than once in a new sample). This 
way, we create bootstrapped datasets that can all be evaluated as a distribution using statistics (mean, median, etc) 
to estimate the data and statistics of the population.

What is the main purpose of bootstrapping?
-Boostrapping is used in order to estimate the statistics of a population provided that we have a sample size n from this
population. It helps us get information from a population meanwhile avoiding us from incurring the heavy costs that 
would come if we were to conduct experiments using the data of the entire population and get their statistics.

If you had a (hypothesized) guess about what the average of a population was, and you had a sample of size n from that
population, how could you use bootstrapping to assess whether or not your (hypothesized) guess might be plausible?
-Using a sample size n, I would use bootstrapping to estimate the average of a population in the following ways: First, I
create a bunch of other samples by resampling the original sample. To do so, I make use of sampling with replacement to 
help me use different data values multiple times in my new datasets. After having created multiple bootstrapped datasets,
I calculate all their means and visually represent these means on a plot (preferably a histogram, KDE or even dot plot). I 
can then easily evaluate the mean of all these bootstrapped means to estimate the average/the mean of my population.
```

```python
import numpy as np

# Data
final_health_scores = np.array([86, 86, 80, 86, 84, 86, 86, 82, 83, 84])

# Bootstrap parameters
n_iterations = 1000  # Number of bootstrap samples
n_size = len(final_health_scores)  # Size of each bootstrap sample

# Generate bootstrap samples
bootstrap_samples = np.random.choice(final_health_scores, (n_iterations, n_size), replace=True)

# Calculate the mean of each bootstrap sample
bootstrap_means = np.mean(bootstrap_samples, axis=1)

# Observed mean
observed_mean = np.mean(final_health_scores)

# Null hypothesis mean
null_hypothesis_mean = 85

# Calculate p-value
p_value = np.sum(np.abs(bootstrap_means - null_hypothesis_mean) >= np.abs(observed_mean - null_hypothesis_mean)) / n_iterations

print(f"Observed Mean: {observed_mean}")
print(f"Bootstrap Mean: {np.mean(bootstrap_means)}")
print(f"p-value: {p_value}")

# Determine if we reject the null hypothesis
alpha = 0.05  # Significance level
if p_value < alpha:
    print("Reject the null hypothesis")
else:
    print("Fail to reject the null hypothesis")
```
